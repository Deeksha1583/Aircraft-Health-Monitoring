```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GroupShuffleSplit
from sklearn.metrics import mean_absolute_error, r2_score
from sklearn.ensemble import RandomForestRegressor
import xgboost as xgb
import joblib
```

```python
train = pd.read_csv(r"Dataset/PM_train.csv")
test = pd.read_csv(r"Dataset/PM_test.csv")
truth = pd.read_csv(r"Dataset/PM_truth.csv")
print(train.head())
print("Train shape: ",train.shape)
```

```python
# Creating RUL {Remaining Useful Life}
max_cycle = train.groupby('id')['cycle'].max().reset_index()
max_cycle.columns = ['id','max_cycle']

train = train.merge(max_cycle, on='id')
train['RUL'] = train['max_cycle'] - train['cycle']
train.drop('max_cycle', axis=1, inplace=True)
print("\nDataset with RUL:")
print(train.head())

# RUL Cap
train['RUL'] = train['RUL'].clip(upper=100)
```

```python
# Data cleaning (removing flat sensor readings)
sensor_cols = [col for col in train.columns if col.startswith('s')]

sensor_std = train[sensor_cols].std()
print("\nSensor Standard Deviations")
print(sensor_std)

flat_sensors = sensor_std[sensor_std < 0.01].index.tolist()
print("Flat Sensors Removed:", flat_sensors)

train.drop(columns=flat_sensors, inplace=True)
```

```python
# Feature Engineering
sensor_cols = [col for col in train.columns if col.startswith('s')]

# Rolling Features (better window)
for col in sensor_cols:
    train[col+"_mean"] = train.groupby('id')[col].rolling(15).mean().reset_index(0,drop=True)
    train[col+"_std"] = train.groupby('id')[col].rolling(15).std().reset_index(0,drop=True)
    train[col+"_min"] = train.groupby('id')[col].rolling(15).min().reset_index(0,drop=True)
    train[col+"_max"] = train.groupby('id')[col].rolling(15).max().reset_index(0,drop=True)
    train[col+"_lag1"] = train.groupby('id')[col].shift(1)
    train[col+"_trend"] = train[col] - train[col+"_lag1"]
    train[col+"_ema"] = train.groupby('id')[col].transform(lambda x: x.ewm(span=10).mean())

max_cycle = train.groupby('id')['cycle'].transform('max')

train.bfill(inplace=True)
```

```python
# Define features
X = train.drop(columns=['id', 'cycle', 'RUL'])
y = train['RUL']
groups = train['id']
```

```python
# Group based Train-test split
gss = GroupShuffleSplit(test_size=0.2, n_splits=1, random_state=42)

for train_idx, val_idx in gss.split(X, y, groups):
    X_train = X.iloc[train_idx]
    X_val = X.iloc[val_idx]
    y_train = y.iloc[train_idx]
    y_val = y.iloc[val_idx]
```

```python
weights = np.where(y_train < 30, 3, 1)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
```

```python
# Random Forest model
rf = RandomForestRegressor(
    n_estimators=300,
    max_depth=20,
    random_state=42,
    n_jobs=-1
)
rf.fit(X_train_scaled, y_train, sample_weight=weights)
rf_pred = rf.predict(X_val_scaled)
```

```python
# XGBoost model
xgb_model = xgb.XGBRegressor(
    n_estimators=1000,
    learning_rate=0.02,
    max_depth=8,
    subsample=0.9,
    colsample_bytree=0.9,
    gamma=0.03,
    reg_lambda=3,
    reg_alpha=1,
    random_state=42,
    n_jobs=-1
)

xgb_model.fit(
    X_train_scaled, y_train, sample_weight=weights
)
xgb_pred = xgb_model.predict(X_val_scaled)
```

```python
final_pred = (0.5 * rf_pred) + (0.5 * xgb_pred)
print(final_pred)
```

```python
# Evaluation
rmse = np.sqrt(np.mean((y_val - final_pred)**2))
print("\nFINAL MODEL PERFORMANCE")
print("MAE:", mean_absolute_error(y_val, final_pred))
print("RMSE:", rmse)
print("R2:", r2_score(y_val, final_pred))
```

```python
joblib.dump(rf, "rf_model.pkl")
joblib.dump(xgb_model, "xgb_model.pkl")
joblib.dump(scaler, "scaler.pkl")
joblib.dump(X.columns.tolist(), "feature_columns.pkl")
print("\nModels saved successfully!")
```
